```python
import pandas as pd
import numpy as np
import tensorflow
import keras
```

```python
df = pd.read_csv('final_dataset.csv')
df.head()
```

```python
X = df.drop(columns=['Yield_tons_per_hectare'])
y = df['Yield_tons_per_hectare']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X, y, train_size=0.8, random_state=30)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
cols = ['Rainfall_mm', 'Temperature_Celsius']
x_train[cols] = sc.fit_transform(x_train[cols])
x_test[cols] = sc.transform(x_test[cols])
```

```python
from keras.models import Sequential
from keras.layers import Dense
ann = Sequential()
ann.add(Dense(input_dim =5 , units =9 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =9 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =1 ,kernel_initializer ='uniform', activation= 'linear'))
ann.compile(optimizer='adam',loss='mean_squared_error')
```

```python
ann.fit(x_train,y_train,batch_size=45,epochs=7)
```

```python
x_train_pred = ann.predict(x_train)
x_test_pred = ann.predict(x_test)
```

```python
#Checking RMSE
from sklearn.metrics import mean_squared_error
Train_MSE=mean_squared_error(y_train,x_train_pred)
Test_MSE=mean_squared_error(y_test,x_test_pred)
print('Train RMSE :',Train_MSE**0.5)
print('Test RMSE :',Test_MSE**0.5)
```
