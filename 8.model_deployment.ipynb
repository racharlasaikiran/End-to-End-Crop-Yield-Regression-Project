```python
# Loading model

import joblib
import numpy as np
import pandas as pd
model = joblib.load('R_model.joblib')
scaler = joblib.load('scaler.joblib')
```

```python
# Collecting future data
Rainfall_mm = [float(input("Rainfall_mm : "))]
Temperature_Celsius = [float(input("Temperature_Celsius : "))]
Fertilizer_Used  = [float(input("Fertilizer_Used : "))]
Irrigation_Used= [float(input("Irrigation_Used : "))]
Days_to_Harvest= [float(input("Days_to_Harvest : "))]
Region_North = [float(input("Region_North : "))]
Region_South = [float(input("Region_South : "))]
Region_West  = [float(input("Region_West : "))]
Soil_Type_Clay= [float(input("Soil_Type_Clay : "))]
Soil_Type_Loam = [float(input("Soil_Type_Loam : "))]
Soil_Type_Peaty = [float(input("Soil_Type_Peaty : "))]
Soil_Type_Sandy  = [float(input("Soil_Type_Sandy : "))]
Soil_Type_Silt= [float(input("Soil_Type_Silt : "))]
Crop_Cotton= [float(input("Crop_Cotton : "))]
Crop_Maize = [float(input("Crop_Maize : "))]
Crop_Rice = [float(input("Crop_Rice : "))]
Crop_Soybean  = [float(input("Crop_Soybean : "))]
Crop_Wheat= [float(input("Crop_Wheat : "))]
Weather_Condition_Rainy= [float(input("Weather_Condition_Rainy : "))]
Weather_Condition_Sunny= [float(input("Weather_Condition_Sunny : "))]

df = pd.DataFrame({'Rainfall_mm':Rainfall_mm, 'Temperature_Celsius':Temperature_Celsius,
                   'Fertilizer_Used': Fertilizer_Used,'Irrigation_Used': Irrigation_Used,
                   'Days_to_Harvest':Days_to_Harvest,'Region_North':Region_North,
                   'Region_South':Region_South,'Region_West':Region_West,
                   'Soil_Type_Clay': Soil_Type_Clay,'Soil_Type_Loam':Soil_Type_Loam,
                   'Soil_Type_Peaty':Soil_Type_Peaty,'Soil_Type_Sandy':Soil_Type_Sandy,
                   'Soil_Type_Silt':Soil_Type_Silt,'Crop_Cotton':Crop_Cotton,
                   'Crop_Maize':Crop_Maize,'Crop_Rice':Crop_Rice,'Crop_Soybean':Crop_Soybean,
                   'Crop_Wheat':Crop_Wheat,'Weather_Condition_Rainy':Weather_Condition_Rainy,
                   'Weather_Condition_Sunny':Weather_Condition_Sunny })
```

```python
df=df.drop(columns=['Soil_Type_Loam'])
df=df.drop(columns=['Weather_Condition_Sunny'])
df=df.drop(columns=['Crop_Rice'])
df=df.drop(columns=['Soil_Type_Peaty'])
df=df.drop(columns=['Region_South'])
df=df.drop(columns=['Region_West'])
df=df.drop(columns=['Crop_Soybean'])
df=df.drop(columns=['Crop_Cotton'])
df=df.drop(columns=['Soil_Type_Silt'])
df=df.drop(columns=['Weather_Condition_Rainy'])
df=df.drop(columns=['Soil_Type_Sandy'])
df=df.drop(columns=['Days_to_Harvest'])
df=df.drop(columns=['Region_North'])
df=df.drop(columns=['Crop_Wheat'])
df=df.drop(columns=['Crop_Maize'])
```

```python
cols = ['Rainfall_mm', 'Temperature_Celsius']
df[cols]= scaler.transform(df[cols])
prediction = model.predict(df)
print("Predicted Yield_tons_per_hectare:", prediction)
```
